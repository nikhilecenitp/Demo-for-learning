```python
import cv2
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
img1= cv2.imread('20200313_164610.jpg')
```

```python
print(img1.shape)
```

```python
#resize the images
img1= cv2.resize(img1,(200,200))
```

```python
#bgr to b/w images
img1= cv2.cvtColor(img1,cv2.COLOR_BGR2BGRA)
```

```python
plt.imshow(img1)
plt.show()
```

```python
edges = cv2.Canny(img1,300,351)
plt.imshow(edges)
plt.show()
```

```python
cv2.imwrite('canny_output.jpg',edges)
```
