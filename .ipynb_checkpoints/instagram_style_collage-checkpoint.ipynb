```python
df1= pd.DataFrame(img[:,:,0])  #r
df2= pd.DataFrame(img[:,:,1])  #g
df3= pd.DataFrame(img[:,:,2])  #b
```

```python
df1.head()
```

```python
# splitting dataframe by row index 
df11= pd.DataFrame()
for i in range(0,430):
    df11=pd.concat([df11, df1.iloc[i,:]]).astype(int)

df22= pd.DataFrame()
for i in range(0,430):
    df22=pd.concat([df22, df2.iloc[i,:]]).astype(int)
    
df33= pd.DataFrame()
for i in range(0,430):
    df33=pd.concat([df33, df3.iloc[i,:]]).astype(int)
    
df11
```

```python
data1= df11[0].tolist()
data2= df22[0].tolist()
data3= df33[0].tolist()
#data1
```

```python
data ={'r':data1, 
       'g':data2, 
       'b':data3} 
  
df = pd.DataFrame.from_dict(data)  
df 
```

```python
df = df[['r','g','b']]
```
